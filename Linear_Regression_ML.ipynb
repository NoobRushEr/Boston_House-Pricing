```python
## Dependent and independent features
x = data.iloc[:, :-1]
y = data.iloc[:, -1]
```

```python
x.head()
```

```python
y.head()
```

```python
## Train test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
x_train
```

```python
y_test
```

```python
## Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler(with_mean=False, with_std=False)
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
import pickle
pickle.dump(sc, open('scaling.pkl', 'wb'))
```

```python
x_train
```

```python
x_test
```

## Model Training

```python
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()
```

```python
regression.fit(x_train, y_train)
```

```python
## print Coefficients and the intercepts
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
regression.get_params()
```

```python
## prediction
y_pred = regression.predict(x_test)
```

```python
y_pred
```

## Assumptions

```python
plt.scatter(y_test, y_pred)
```

```python
residuals = y_test - y_pred
```

```python
sns.displot(residuals, kind='kde')
```

```python
plt.scatter(y_pred, residuals)
```

## Model Evaluation

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
print(mean_absolute_error(y_test, y_pred))
print(mean_squared_error(y_test, y_pred))
print(np.sqrt(mean_squared_error(y_test, y_pred)))
```

## R square and adjusted R square

Formula
##### R^2 = 1 - SSR/SST
R^2 = Coefficient of determination of SSR = Sum of squares of residuals SST = Total sum of squares

```python
from sklearn.metrics import r2_score
score = r2_score(y_test, y_pred)
score
```

##### Adjusted R^2 = [(1 - R^2) * (n - 1) / (n-k-1)]
Where:
R2: The R2 of the model n:The number of observations k:The number of predictions

```python
1 - (1-score) * (len(y_test) - 1) / (len(y_test) - x_test.shape[1] -1)
```
